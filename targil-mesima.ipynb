### שימוש בנתונים

```python
def getGradesList():
    grades=[]
    f=open("grades.csv")
    listOfStrings=f.readlines()
    for l in listOfStrings[1:]:
        l=l.strip('\n')
        pair=l.split(',')
        grades.append(int(pair[1]))
    return grades

grades=getGradesList()
maximum=max(grades)
minimum=min(grades)
print(maximum)
print(minimum)
```

### חציון וממוצע עם statistics

```python
import statistics

grades=getGradesList()

med=statistics.median(grades)
avg=statistics.mean(grades)

print(med,avg)
```

## רונאלדו

```python
def prepareData(fileName):
    data=[]
    f=open(fileName)
    listOfStrings=f.readlines()
    for l in listOfStrings:
        data.append(l.strip().split(','))
    return data

fileName=r"/home/tal/Downloads/ronaldo_data.csv"
print(prepareData(fileName))
```

```python
data=prepareData(fileName)
print(data[0])
```

### באיזה מועדונים שיחק רונאלדו?

```python
ronaldo=[]
for d in data[1:]:
    r=dict(zip(data[0],d))
    ronaldo.append(r)
    print(r)
    
```

```python
for r in ronaldo:
    print(r["Club"])
```

```python
clubs=set()
for r in ronaldo:
    clubs.add(r["Club"])
print(clubs)
```

### באיזה שנים שיחק רונאלדו בריאל מדריד?

```python
years=[]
for r in ronaldo:
    if r["Club"]=='Real Madrid':
        years.append(r["Date"])
print(years)
```

```python
s=set()
for y in years:
    s.add('20'+y.split('/')[2])
print(s)
print(sorted(list(s)))
years=sorted(list(s))
```

### באיזה שנה היו לרונאלדו הכי הרבה נצחונות ובאיזו קבוצה?

```python
winningYears=[]
for r in ronaldo:
    result=r["Result"].split(':')
    if int(result[0])>int(result[1]):
        winningYears.append(r["Date"])
```
