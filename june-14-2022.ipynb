```python
# better use number of rows 

def make_fake_data_table(rowNum=5,cols=["name", "phone", "email"]):
    dataTable={}
    for i in range(rowNum):
        key=i
        row={i:{}}
        for j in cols:
            row[i][j]=i*100
        dataTable.update(row)
    return dataTable

print(make_fake_data_table())

print(make_fake_data_table()[0])
print(make_fake_data_table()[0]["phone"])

```

```python

# adding some fake data
from faker import Faker
fake=Faker(locale='he_IL')

def make_fake_data_table(rowNum=3,cols=["name", "phone", "email"]):
    dataTable={}
    for i in range(rowNum):
        key=fake.iana_id()
        row={key:{}}
        for j in cols:
            row[key].update(zip(cols, [fake.name(), fake.phone_number(), fake.email()]))
        dataTable.update(row)
    return dataTable

print(make_fake_data_table())
```
