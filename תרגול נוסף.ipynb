### הצעה לפתרון
תקנו או הוסיפו מה שחסר

```python
# guessing game

import random
from sys import exit

randSize=20
turns=10
tries=0
gameMode=""
score=0
bonus=0
topTen=[]
randNum=0

userName=input("What is player name: ")
gameMode=int(input(f"""
Hello, {userName}, welcome to gussing game!
Please choose range of numbers to guess from:
1) 1-10, 20 turns
2) 1-20, 10 turns
3) 1-50, 10 turns
4) 1-100, 8 turns
5) CrAzY, 5 turns
>     """))
if gameMode==1:
    randSize=10
    turns=20
    randNum=random.randint(1,randSize)
if gameMode==2:
    randSize=20
    turns=10
    randNum=random.randint(1,randSize)
if gameMode==3:
    randSize=50
    turns=10
    randNum=random.randint(1,randSize)
if gameMode==4:
    randSize=100
    turns=8
    randNum=random.randint(1,randSize)
if gameMode==5:
    randSize=random.randint(1,100)
    turns=5
    randNum=random.randint(randSize-30,randSize)
    
# game loop

while tries<turns:
    guess=int(input("Guess that number!: "))
    tries+=1
    if guess<randNum:
        print(f"No, your guess is too low...you have {turns-tries} turns left. Try again.")
    if guess>randNum:
        print(f"No, your guess is too high...you have {turns-tries} turns left. Try again.")
    if guess==randNum:
        if tries<3: bonus=10 
        score=turns-tries
        print(f"Correct! Well Done {userName}! Your score is {score}!")
        if bonus>0:
            print(f"You got a bonus of {bonus} points, so your final score is {score+bonus}!")
        topTen.append([score, userName])
        topTen.sort(reverse=True)
        print("\nHere is the top ten list:\n")
        for t in topTen:
            if t==topTen[0]:
                print(f"NUMBER ONE: {t[1]}\t Score {t[0]}")
            else:
                print(f"\t{t[1]}\t Score {t[0]}")
        again=input("Play again? (y/n)")
        if again=='y':
            turns=10 # We are giving 10 more turns, it can depend on score
            randNum=random.randint(1,randSize) # Can increase with each round
            tries=bonus=score=0
            userName=input("Your name: ")
        else:
            print("Good bye!")
            exit()
print("Sorry, you lost...")
        
            
        



```

```python
# 21 Card Game
# See nice solution in https://inventwithpython.com/bigbookpython/project4.html

from random import randint

playerName=""
cards=[*[c for c in range(1,11)], "prince", "queen", "king", "ace"]
HEARTS   = chr(9829) # Character 9829 is '♥'.
DIAMONDS = chr(9830) # Character 9830 is '♦'.
SPADES   = chr(9824) # Character 9824 is '♠'.
CLUBS    = chr(9827) # Character 9827 is '♣'.
types=[HEARTS,DIAMONDS,SPADES,CLUBS]

dealerHand=0
playerHand=0
playerCards=[]
dealerCards=[]

bet=0
score=0

playerCard1=randint(1,len(cards))
playerCard2=randint(1,len(cards))
playerCards+=[playerCard1, playerCard2]
dealerCard1=randint(1,len(cards))
dealerCard2=randint(1,len(cards))
dealerCards+=[dealerCard1, dealerCard2]

while True: # game loop
    print("Your cards are: ", *playerCards, sep='\n')
    dealerHand=sum(dealerCards)
    playerHand=sum(playerCards)
    if playerHand==21:
        print("BLACKJACK!!!")
        score+=100
        continue
    if playerHand>21 or dealerHand==21:
        print(f"you are bust, I win, my cards: {dealerCard1, dealerCard2}")
        break
    if dealerHand>21:
        print(f"you busted the dealer, you win! dealer hand: {dealerCard1, dealerCard2}")
        break
    playerChoice=input(f"""
    Do you want to:
    1) Hit (ask for another card)
    2) Stand (end your turn)
    """)
    if playerChoice=='1':
        playerCards.append(randint(1,len(cards)))
    if playerChoice=='2':
        dealerCards.append(randint(1,len(cards)))
        
```
