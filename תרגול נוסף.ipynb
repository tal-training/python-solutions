## משחק ניחושים 

כתבו תכנית שבוחרת מספר אקראי בין 1 למספר שתגדירו.
בפייתון משתמשים בפונקצייה הבאה כדי לבחור מספר אקראי, נניח בין 1 ל-20:
import random
randNum=random.randint(1,20)
המשתמש צריך לנחש את המספר.
יש למשתמש מספר מוגבל של נסיונות
אחרי כל ניחוש המשתמש יקבל חיווי אם הניחוש גדול או קטן מדי.
אפשר לשפר את המשחק:
הוסיפו ניקוד המתעדכן בין תור לתור
הוסיפו ניקוד בונוס ככל שהניחוש יותר מהיר
הוסיפו אפשרות להגדיר מספר נסיונות וטווח הניחוש
הוסיפו שם משתמש ורשימת שחקנים מובילים
שנו את הניחוש למילים או אותיות במקום מספרים

```python
# Guessing game
import random
from sys import exit

randSize=20
turns=10
tries=0
gameMode=""
score=0
bonus=0
topTen=[]
randNum=0

userName=input("What is player name: ")
gameMode=int(input(f"""
Hello, {userName}, welcome to gussing game!
Please choose range of numbers to guess from:
1) 1-10, 20 turns
2) 1-20, 10 turns
3) 1-50, 10 turns
4) 1-100, 8 turns
5) CrAzY, 5 turns
>     """))
if gameMode==1:
    randSize=10
    turns=20
    randNum=random.randint(1,randSize)
if gameMode==2:
    randSize=20
    turns=10
    randNum=random.randint(1,randSize)
if gameMode==3:
    randSize=50
    turns=10
    randNum=random.randint(1,randSize)
if gameMode==4:
    randSize=100
    turns=8
    randNum=random.randint(1,randSize)
if gameMode==5:
    randSize=random.randint(1,100)
    turns=5
    randNum=random.randint(randSize-30,randSize)
    
# game loop

while tries<turns:
    guess=int(input("Guess that number!: "))
    tries+=1
    if guess<randNum:
        print(f"No, your guess is too low...you have {turns-tries} turns left. Try again.")
    if guess>randNum:
        print(f"No, your guess is too high...you have {turns-tries} turns left. Try again.")
    if guess==randNum:
        if tries<3: bonus=10 
        score=turns-tries
        print(f"Correct! Well Done {userName}! Your score is {score}!")
        if bonus>0:
            print(f"You got a bonus of {bonus} points, so your final score is {score+bonus}!")
        topTen.append([score, userName])
        topTen.sort(reverse=True)
        print("\nHere is the top ten list:\n")
        for t in topTen:
            if t==topTen[0]:
                print(f"NUMBER ONE: {t[1]}\t Score {t[0]}")
            else:
                print(f"\t{t[1]}\t Score {t[0]}")
        again=input("Play again? (y/n)")
        if again=='y':
            turns=10 # We are giving 10 more turns, it can depend on score
            randNum=random.randint(1,randSize) # Can increase with each round
            tries=bonus=score=0
            userName=input("Your name: ")
        else:
            print("Good bye!")
            exit()
print("Sorry, you lost...")
        
            
        



```

## קופה רושמת
כיתבו תוכנית המממשת קופה רושמת. 
הגדירו רשימת מוצרים ומחירים

אחרי שבוחר סיום מקבל חשבון עם רשימת המוצרים שבחר וסך הכל המחיר כולל מעמ
שלבים מתקדמים
הציגו את סהכ החשבון תוך כדי ההזנה
הציגו כמות מכל מוצר שנקנה כולל כפילויות
הציגו אפשרויות לתשלום מזומן או אשראי
במידה ובוחר אשראי יש לוודא שהמספר המוזן הוא בפורמט של כרטיס אשראי
הוסיפו אפשרות של קבלת שם משתמש והנחות לקוח מועדף ושמירת כרטיס אשראי


```python
products={
    "[a]pple":6,
    "[b]anana":12,
    "[c]heese":13,
    "[d]rinks":17,
    "[e]ggs":8
}

menu=""
for k in products.keys():
    menu+=f"{k}: {products[k]} shekel\n"
print(menu)

choices=[]
choice=input("Choose a product, to pay enter '0': ")

while choice!='0':
    for p in products.keys():
        if choice[0]==p[1]:
            choices.append(p)
            print(f"added {p} to list")
    choice=input("Choose a product, to pay enter '0': ")

total=0
for c in choices:
    total+=products[c]

total*=1.17

print(f"""
קבלה עין ההוד סחר בע"מ
-------------------
{[x.replace('[','').replace(']','')+': '+str(products[x]) for x in choices]}
your reciept is total is {total}
""")
        
```
